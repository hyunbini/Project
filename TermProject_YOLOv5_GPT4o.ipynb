```python
# ③ data.yaml 경로 수정 (Google Drive 기준)
yaml_path = "/content/drive/MyDrive/yolov5_data/data.yaml"

# 내부 경로 자동 수정
with open(yaml_path, 'r') as f:
    content = f.read()

content = content.replace("FurnitureDecor-3/train/images", "/content/drive/MyDrive/yolov5_data/train/images")
content = content.replace("FurnitureDecor-3/valid/images", "/content/drive/MyDrive/yolov5_data/valid/images")
content = content.replace("../test/images", "/content/drive/MyDrive/yolov5_data/test/images")

# 새로운 yaml 저장
with open("data.yaml", 'w') as f:
    f.write(content)
```

```python
# ④ custom 하이퍼파라미터 저장
custom_hyp = """
lr0: 0.002
lrf: 0.2
momentum: 0.937
weight_decay: 0.0005
warmup_epochs: 2.0
warmup_momentum: 0.8
warmup_bias_lr: 0.1

box: 0.05
cls: 0.4
obj: 0.7
anchor_t: 4.0
cls_pw: 1.0
obj_pw: 1.0
label_smoothing: 0.0
fl_gamma: 0.0   # ✅ focal loss gamma

hsv_h: 0.02
hsv_s: 0.7
hsv_v: 0.4
degrees: 0.0
translate: 0.1
scale: 0.5
shear: 0.0
perspective: 0.0
flipud: 0.0
fliplr: 0.5
mosaic: 1.0
mixup: 0.1
copy_paste: 0.0
"""
with open("custom_hyp.yaml", "w") as f:
    f.write(custom_hyp)
```

```python
# ⑤ 학습 시작 (15 epoch, batch 16)
!python train.py \
  --img 640 \
  --batch 16 \
  --epochs 15 \
  --data data.yaml \
  --hyp custom_hyp.yaml \
  --weights yolov5s.pt \
  --name tuned_yolo_drive
```
